```python
import random
height,weight , score= [],[],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,200))
    score.append(random.randint(1,100))
    
plt.rc('font',family='malgun gothic')
plt.title('키와 몸무게, 체육 점수의 상관관계')

plt.scatter(height,weight,s= score, c=score ,cmap='rainbow', alpha=0.5)  #  원의 사이즈 - s    c - 색깔의 진하기
plt.colorbar(label='체육점수')
plt.xlabel=('키')
plt.ylabel=('몸무게')

plt.show()
```

```python
import matplotlib.pyplot as plt
singer=['A','B',"C","D","E"]
week1=[42,58,19,92,84]
week2= [53,52,48,98,73]

for i in range(len(week1)):
    week1[i]= -week1[i]         # 왼쪽으로 표현되도록  음수화 함.

plt.rcParams['axes.unicode_minus']=False   # - 를 제대로 표시하기 위한 구문
plt.rc('font',family='malgun gothic')               # 한글을 제대로 표시하기 위한 구문
plt.title('오디션 프로그램 득표 현황')

plt.barh(singer,week1, label='첫째 주', color='hotpink')
plt.barh(singer,week2, label='둘째 주', color='royalblue')
plt.legend()
plt.show()
```
