```python
## 여의도 한강공원
## 여의도 공원
# 샛강생태공원 

lat1, long1 = 37.52862, 126.9331 # 여의도 한강공원
lat2, long2 = 37.52565, 126.9215 # 여의도공원
lat3 ,long3 = 37.51888, 126.9219 # 샛강생태공원

map_y = folium.Map([lat1, long1], zoom_start=15)
folium.Marker([lat1, long1], tooltip='여의도 한강공원', icon=folium.Icon(color='red', icon='heart', prefix='fa')).add_to(map_y)
folium.Marker([lat2, long2], tooltip='여의도공원', icon=folium.Icon(color='blue', icon='home', prefix='fa')).add_to(map_y)
folium.Marker([lat3 , long3], tooltip='샛강생태공원', icon=folium.Icon(color='green', icon='flag', prefix='fa')).add_to(map_y)
map_y


### 설정 가능한 색상
### 'red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 
### 'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white', 
### 'pink', 'lightblue', 'lightgreen', 'gray', 'black', 'lightgray'


### 다양한 마커 아이콘
### https://fontawesome.com/v4.7.0/icons
### fontawesome의 아이콘을 사용할 때에는 prefix='fa' 옵션을 추가

```

```python
lat = [37.52862, 37.52565, 37.51888]                         # 위도 리스트  
long = [ 126.9331, 126.9215, 126.9219]                       # 경도 리스트
names = ['여의도 한강공원', '여의도공원', '샛강생태공원']    # 위치 이름 리스트
icons = ['automobile', 'balance-scale', 'ban']               # 아이콘 모양 리스트   
colors = ['red', 'blue', 'purple']                           # 아이콘 색상 리스트

map_y = folium.Map([lat[0], long[0]], zoom_start = 15)

for i in range(len(lat)) : 
    folium.Marker([lat[i], long[i]], tooltip=names[i],
                  icon=folium.Icon(color=colors[i], icon=icons[i], prefix='fa')).add_to(map_y)

map_y
```

```python
map_y = folium.Map([35.15961, 129.1606], zoom_start = 15)

folium.CircleMarker([35.15961, 129.1606], 
                    tooltip='부산 해운대 해수욕장',
                    radius=100,         #원의 반지름
                    color='brown',         #원의 둘레 색상
                    fill=True,
                    fill_color='coral', #원의 안쪽 색상
                    fill_opacity=0.5      #투명도
                   ).add_to(map_y)

map_y
```

```python
import folium

lat = [35.15870, 35.23360, 35.17357, 35.1734932]
long = [129.16047, 129.08309, 128.94632, 129.1819789]
names = ['해운대 해수욕장', '부산대학교', '김해국제공항', '해운대 백병원'] 

icons = ['hand-o-down', 'university', 'plane', 'ambulance']                 
colors = ['red', 'blue', 'purple','green']                               

map_h = folium.Map([lat[0], long[0]],tiles='Stamen Watercolor', zoom_start = 11)

### ,tiles='Stamen Toner'
### ,tiles='Stamen Terrain'


for i in range(len(lat)) : 
    folium.Marker([lat[i], long[i]], tooltip=names[i],
                  icon=folium.Icon(color=colors[i], icon=icons[i], prefix='fa')).add_to(map_h)

map_h
```
