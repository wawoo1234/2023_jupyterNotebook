```python
import numpy as np

data = np.loadtxt('D:/2023_jsm_r/jupyter_2023_j/data/chocolate_rating.csv',delimiter = ',' )

print('차원:', data.ndim,'\n')

print('모양:', data.shape,'\n')

print('원소 수 : ', data.size,'\n')

print(data)
```

```python
# 모든 초콜릿의 평균 평점

ratings_mean = data[:,3]. mean()    ##  넘파이에서는 : 이 모든을 뜻함. 
print(ratings_mean)

```

```python
# 평점이 4 이상인 우수 초콜릿을 high_level 변수로 저장

high_level = data [ data[:,3]>=4]
   
print(high_level)


```

```python
# 우수초콜릿(high_level)의 번호를 정수로 변환해서 high_id 변수로 저장 

high_id = high_level[:,0].astype(np.int64)   # 0번 인덱스 타입을 정수형으로 변환해서 우수 초콜릿의 개수
print(high_id)

print('\n우수  초콜릿 개수 : ', high_id.size)
```
