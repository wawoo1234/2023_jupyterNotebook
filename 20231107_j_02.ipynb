```python
df = pd.read_csv('./data/기온.csv', encoding='cp949')      # 주피터 노트북 경로 안에 데이터 폴더 안에 파일 넣어두기      
                                                                                                    # 최근 encoding이cp949가 많음. 안되면 예전 utf-8이 
df
```

```python
df.info()                       # 기본 정보 확인하기       # 컬럼 항목의 개수 8개 ,,, 0~7 항목 이름들    
                                                                                    # 널값이 아닌 값들을 카운트한 값    - 결측값 확인하기   # 최저기온에 결측값 있음을 확인함. 
                                                                                    # object - 문자열 의미 
                                                                                    #  float64(항목이 4), int64(인티저값 항목이 2개 ), object(항목이 2개)
```

```python
df2 = df.fillna(method = 'ffill')   # ffill의 의미는  forward fill', 누락된 값을 이전 값으로 채움
                                                       # bfill의 의미는: 누락된 값을 뒤쪽 값으로 채움
    
df2.info()
df2.to_csv('./data/기온_2.csv',encoding='cp949')             # 결측값을 처리한 파일 - csv 파일로 내보내기 
```

```python
df2.rename(columns={'최저기온(°C)': 'min_temp'},  inplace=True)
df2.rename(columns={'평균기온(°C)': 'avg_temp'},   inplace=True)
df2.rename(columns={'최고기온(°C)': 'max_temp'}, inplace=True)
df2.head(3)
```

```python
plt.title('서울시 2022년도 기온 변화')
plt.plot(range(1,len(df2)+1), df2['max_temp'], label='최고기온', c= 'r')
plt.plot(range(1,len(df2)+1), df2['avg_temp'], label='평균기온', c= 'y')
plt.plot(range(1,len(df2)+1), df2['min_temp'], label='최저기온', c= 'b')
plt.xlabel('일')
plt.ylabel('기온')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family='Malgun Gothic')               # 한글 처리를 위함 - matplotlib
plt.rcParams['axes.unicode_minus'] = False            # 축의 음수 기호 표시를 위함.

def plot_monthly_temperatures(df, month, ax):
    df_month = df[df['일시'].dt.month == month]

    ax.set_title(f'서울시 2022년도 {month}월 기온 변화')
    ax.plot(range(1, 32 if month in [3, 12] else 31), df_month['max_temp'], label='최고기온', c='r')
    ax.plot(range(1, 32 if month in [3, 12] else 31), df_month['avg_temp'], label='평균기온', c='y')
    ax.plot(range(1, 32 if month in [3, 12] else 31), df_month['min_temp'], label='최저기온', c='b')
    ax.set_xlabel(f'{month}월')
    ax.set_ylabel('기온')
    ax.set_xlim(1, 32 if month in [3, 12] else 31)
    ax.set_ylim(-15, 30)

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

months = [3, 4, 11, 12]

for i, month in enumerate(months):
    plot_monthly_temperatures(df2, month, axs[i // 2, i % 2])

plt.legend()
plt.tight_layout()


plt.savefig('서울.png')     # plt.savefig  함수  - 결과를 이미지로 저장함.

# plt.show()
```
