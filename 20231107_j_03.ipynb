## 결측치 처리 - 평균값으로 대체

```python
# 결측치 평균대체를 위해 평균 확인

air.mean()      # 결측치가 같이 있어서 경고 메시지 뜸
```

```python
# 평균값으로 결측치 대체

air_m = air.fillna(air.mean())
air_m 
```

```python
air_dd=air.set_index('MMDDHH')
air_dd.mean()
```

air_dd=air.set_index('MMDDHH'): 
이 부분은 데이터프레임 air에서 'MMDDHH' 열을 새로운 인덱스로 설정하고, 
결과로 새로운 데이터프레임 air_dd를 생성합니다. 
즉, 'MMDDHH' 열은 이제 데이터프레임의 인덱스로 사용됩니다.

air_dd.mean(): 이 부분은 새로운 데이터프레임 air_dd에 대해 열 별로 평균을 계산합니다. 
데이터프레임의 열은 기존 'MMDDHH' 열을 제외한 다른 열들로 구성되며, 
이러한 열들에 대한 각각의 열 별 평균을 반환합니다.

결과적으로 air_dd.mean()은 데이터프레임 air_dd의 각 열에 대한 평균값을 나타내며, 
이를 시리즈(Series) 형태로 반환합니다. 이를 통해 각 열의 평균값을 확인할 수 있습니다.

```python
# 평균값으로 결측치 대체

air_m = air.fillna(air_dd.mean())
air_m
```

# 중앙값으로 결측치 대체

```python
air_dd.median()    #중앙값
```

```python
air_md = air.fillna(air_dd.median())
air_md
```

주어진 코드는 Pandas 데이터프레임에서 
결측값(누락된 값)을 해당 열의 중앙값(중앙값)으로 대체하는 작업을 수행합니다. 

air_md = air.fillna(air_dd.median()): 
    이 부분은 데이터프레임 air에서 결측값을 대체하는 작업을 수행합니다.
    
air_dd.median(): 
    air_dd 데이터프레임의 각 열에 대해 중앙값을 계산합니다. 
    중앙값은 데이터의 중간값으로, 결측값을 제외한 값들 중에서 중간에 위치한 값을 의미합니다.
    
fillna(): 결측값을 대체하는 Pandas 메서드입니다. 
    여기에서는 데이터프레임 air에서 결측값을 air_dd.median()로 대체합니다.
    
결과적으로 air_md 데이터프레임은 원래 air 데이터프레임과 동일한 구조를 가지며, 
결측값이 중앙값으로 대체된 상태로 저장됩니다. 

이렇게 결측값을 대체하면 데이터의 완전성을 유지하고 
분석이나 시각화 작업을 수행할 때 누락된 데이터로 인한 오류를 방지할 수 있습니다.
