2. 'chick'항목의 합계화 평균, 표준편차, 중앙값, 최소값, 최대값 구하기 

```python
hat_sum = hat['chick'].sum()
print(f'합계: {hat_sum}')

hat_mean = hat['chick'].mean()
print(f'평균: {hat_mean}')

hat_std = hat['chick'].std()
print(f'표준편차: {hat_std}')

hat_med = hat['chick'].median()
print(f'중앙값: {hat_med}')

hat_min = hat['chick'].min()
print(f'최소값: {hat_min}')

hat_max = hat['chick'].max()
print(f'합계: {hat_max}')
```

```python
import seaborn as sns

col7=sns.color_palette('Pastel2',7)   #  Pastel2 색상에서 7가지 색상 가져와서 함. 

plt.figure(figsize=(30,17))

plt.bar(hat['hatchery'], hat['chick'], color = col7, edgecolor = 'black')

plt.title('부화장별 병아리 부화현황', fontsize=17)
plt.xlabel('부화장')
plt.ylabel('부화마릿수')

plt.show()
```

```python
def addtext(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]+0.5, y[i], ha='center')
```

```python
# 막대 위에 텍스트 추가하기
col7 = sns.color_palette('Pastel2', 7)
plt.figure(figsize=(10, 7))

plt.bar(hat['hatchery'], hat['chick'], color = col7, edgecolor = 'black')

addtext(hat['hatchery'], hat['chick'])       # 텍스트 표시 사용자 정의 함수 추가

plt.title('부화장별 병아리 부화현황', fontsize =17)
plt.xlabel('부화장')
plt.ylabel('부화마릿수')
plt.show()
```

```python
# 막대 위에 텍스트 추가하기
def addtext(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]-0.5, y[i], ha='center')

col7 = sns.color_palette('Pastel2', 7)
plt.figure(figsize=(10, 7))

plt.bar(hat['hatchery'], hat['chick'], color = col7, edgecolor = 'black')

addtext(hat['hatchery'], hat['chick'])       # 텍스트 표시 사용자 정의 함수 추가

plt.title('부화장별 병아리 부화현황', fontsize =17)
plt.xlabel('부화장')
plt.ylabel('부화마릿수')
plt.show()
```

```python
# 막대 위에 텍스트 추가하기
def addtext(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]/2, y[i], ha='center')

col7 = sns.color_palette('Pastel2', 7)
plt.figure(figsize=(10, 7))

plt.bar(hat['hatchery'], hat['chick'], color = col7, edgecolor = 'black')

addtext(hat['hatchery'], hat['chick'])       # 텍스트 표시 사용자 정의 함수 추가

plt.title('부화장별 병아리 부화현황', fontsize =17)
plt.xlabel('부화장')
plt.ylabel('부화마릿수')
plt.show()
```
