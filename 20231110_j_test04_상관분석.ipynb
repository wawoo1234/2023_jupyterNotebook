```python
w.info()
```

```python
w2 = w.iloc[:,1:5]
w2.head()
```

```python
w_cor = w2.corr(method = 'pearson')  ## correlation (상관계수)
w_cor
```

```python
## 상관분석 결과 표현하기 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 산점도 그리기

sns.pairplot(w2)
```

```python
sns.pairplot(w2,hue='weight')          # 색깔로 hue에 넣은 'weight' 부분을 표시함.    
```

```python
# 상관행렬도 그리기
plt.figure(figsize = (10,7))
sns.heatmap(
    w_cor, 
    annot = True,             # 각 칸에 수치들을 나타낼지 설정
    cmap = 'Blues',          # color Map
    cbar=True                   # 색상막대 표시 
           )

plt.show()
```
